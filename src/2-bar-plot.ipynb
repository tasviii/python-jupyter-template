### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['blue', 'green', 'red', 'purple']

if not categories or not values:
    raise ValueError("Categories and values must not be empty.")

if len(categories) != len(values):
    raise ValueError("Categories and values lists must be of the same length.")

if len(colors) != len(categories):
    raise ValueError("The number of colors must match the number of categories.")

plt.bar(categories, values, color=colors)

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title("Category Values Bar Plot")

plt.show()

```
