### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(1, 10, 10)
y = x ** 2

categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

scatter_x = np.random.rand(20) * 10
scatter_y = np.random.rand(20) * 100

pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

if x.size == 0 or y.size == 0:
    raise ValueError("x and y data must not be empty.")
if len(x) != len(y):
    raise ValueError("x and y lists must be of the same length.")

if len(categories) != len(values):
    raise ValueError("Categories and values lists must be of the same length.")

if scatter_x.size == 0 or scatter_y.size == 0:
    raise ValueError("Scatter plot data (x and y) must not be empty.")
if len(scatter_x) != len(scatter_y):
    raise ValueError("Scatter plot x and y lists must be of the same length.")

if len(pie_labels) != len(pie_sizes):
    raise ValueError("Pie chart labels and sizes lists must be of the same length.")

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line chart
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axs[0, 0].set_title("Line Chart")

# Bar chart
axs[0, 1].bar(categories, values, color='g')
axs[0, 1].set_title("Bar Chart")

# Scatter plot
axs[1, 0].scatter(scatter_x, scatter_y, color='r')
axs[1, 0].set_title("Scatter Plot")

# Pie chart
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'orange'])
axs[1, 1].set_title("Pie Chart")

plt.tight_layout()

plt.show()



```
