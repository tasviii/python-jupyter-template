### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt

def plot_histogram(data):
    if not isinstance(data, list):
        raise ValueError("Input data must be a list.")
    if not all(isinstance(i, (int, float)) for i in data):
        raise ValueError("All elements in the data list must be numbers.")
    if len(data) == 0:
        raise ValueError("Data list cannot be empty.")

   
    plt.hist(data, bins=len(set(data)), color='purple', edgecolor='black')

   
    plt.title("Histogram of Data")
    plt.xlabel("Value")
    plt.ylabel("Frequency")

    plt.show()


data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

try:
    plot_histogram(data)
except ValueError as e:
    print(f"Error: {e}")
```
